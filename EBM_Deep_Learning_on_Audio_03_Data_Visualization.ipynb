# Deep Learning in Audio Classification in Python

## Audio Data Visualization

```python
import numpy as np
import scipy.io.wavfile
```

```python
sample_rate, signal = scipy.io.wavfile.read('content/COVID19_Cough.wav')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(18, 5))

plt.plot(signal)
plt.show()
```

### Zooming into the Specific Zone

```python
plt.figure(figsize=(18, 5))

plt.title("COVID'19 Cough")
plt.ylabel("Amplitude")
plt.xlabel("Time")

Sample = signal[5500:6000]

plt.plot(Sample)
plt.show()
```

## Time Series Plot

```python
Time = np.linspace(0, len(signal) / sample_rate, num=len(signal))
```

```python
plt.figure(figsize=(18, 5))

plt.title("COVID'19 Cough")
plt.ylabel("Amplitude")
plt.xlabel("Time")

plt.plot(Time, signal)
plt.show()
```
